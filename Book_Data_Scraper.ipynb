```python
error_list=[]
```

```python
def get_data(browser,query):
    start="https://www.google.com/search?tbm=bks&q="
    safe_string = start+urllib.parse.quote_plus(query)

    browser.get(safe_string)
    captcha=[1,4,4]
    while(len(captcha)>0):
        captcha=browser.find_elements(By.CLASS_NAME,"rc-anchor-center-item.rc-anchor-checkbox-label")
        if(len(captcha)>0):
            if(captcha.text=="I'm not a robot"):
                browser.get(safe_string)
        
    
    search=browser.find_element(By.CLASS_NAME,"LC20lb.MBeuO.DKV0Md")
    search.click()
 
    close_list=browser.find_elements(By.CLASS_NAME,"kt12b")
    close=close_list[0]
    try:
        close.click()
    except:
        pass
    
    try:
        expand=browser.find_element(By.CLASS_NAME,"LjDke TKZwpe cjSy4c p8Ujvb bZm3Db".replace(" ","."))
        expand.click()
    except:
        pass
    
    data=""

    datas=browser.find_elements(By.CLASS_NAME,"Y0Qrof")
    for i in datas:
        data+=i.text+"\n"
  
    return data
#     return final[:-10]


```

```python

df=pd.read_csv("popular.csv")
df.fillna("")


df=df.drop(df.columns[0],axis=1)
df
```

```python
q=queue.Queue()


for i in df.index:
    q.put(i)
    
```

```python

def thread(ind):
    global q
    global drivers
    while not q.empty():
        i=q.get()
        try:
            content=get_data(drivers[ind],df.iloc[i,3] + " ," + df.iloc[i,4])
            df.iloc[i,10]=content
            print(q.qsize())
        except:
            print("Error:",i)
            error_list.append(i)
```

# Threading 

```python
for i in range(8):
    threading.Thread(target=thread,args=(i,)).start()
```

# Convert to csv
